# Boston Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow as tf
```

```python
print(tf.__version__)
```

```python
from tensorflow import keras
```

```python
from sklearn.datasets import fetch_openml

boston = fetch_openml(name='boston', version=1, as_frame=True)
X, y = boston.data, boston.target
```

```python
print(X.shape, y.shape)
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(y)
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

# Deep Learning - MultiLayer Perceptron Model (Neural Network)

```python
# Sequential API
dnn = tf.keras.models.Sequential()
```

```python
# Adding first hidden layer

dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))

# Adding second hiiden layer
dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))

# Adding output layer
dnn.add(tf.keras.layers.Dense(units=1))
```

```python
# compiling the network with optimization method
dnn.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# trained your model with input dataset
dnn.fit(x_train, y_train, batch_size=32, epochs=100)
```

```python
dnn.fit(x_train, y_train, batch_size=32, epochs=500)
```

```python
# wanted to predict price for the x_test data
y_pred = dnn.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
# 1. r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# 2. Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")
```

```python
# 3. Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse}")
```

```python
# 4. Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
y_pred = np.array([str(x).strip() for x in y_pred.flatten()])
```

```python
# 5. Residuals Analysis (Plotting)
residuals = y_test - y_pred
plt.figure(figsize=(10, 6))
```

```python
# Plot residuals
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.show()
```

```python
# 6. Plot Actual vs Predicted
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2)  # Ideal line
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted')
plt.show()
```
